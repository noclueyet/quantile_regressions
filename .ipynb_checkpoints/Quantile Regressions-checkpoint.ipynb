```python
# First, lets define the input, prediction, and expected values
x_in = x_0[0]
y_out = y_0[0]
yhat_out = yhat[0]
```

```python
# Estimate stdv in prediction direction
def calc_stdev(y_0, yhat):
    sum_errs = np.sum((y_0-yhat)**2)
    stdev = np.sqrt(1/(len(y_0)-2) * sum_errs) 
    return stdev
stdev = calc_stdev(y_0, yhat)
```

```python
# calculate prediction interval
def pred_interval(stdev, y_out):
    interval = 1.96 * stdev
    return interval
interval = pred_interval(stdev, y_out)
```

```python
# plot
plt.scatter(x_0, y_0)
plt.plot(x_0, yhat, color="r")
plt.errorbar(x_in, yhat_out, yerr=interval, color="black", fmt="o")
```

```python
# calculate prediction interval
interval = 1.96 * stdev
print('Prediction Interval: %.3f' % interval)
lower, upper = y_out - interval, y_out + interval
print('95%% likelihood that the true value is between %.3f and %.3f' % (lower, upper))
print('True value: %.3f' % yhat_out)
```

```python
# plot
plt.scatter(x_0, y_0)
plt.plot(x_0, yhat, color="r")
for i in range(len(yhat)):
    plt.errorbar(x_0[i], yhat[i], yerr=pred_interval(calc_stdev(y_0, yhat),y_0[i]),
                 color="b", fmt="o")
```
